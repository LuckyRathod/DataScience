## T22 - Univariate,Bivariate,Multivariate Analysis (EDA) - Theory

It will help us to do analysis on features of dataset.

#### Univaraite Analysis

In univariate analysis we take One feature and based on these one feature we will try to determine what will be the output.
If we take weight as one features then it will be plotted on x axis.On y axis , all values will be 0.These univaraite analysis will also help us to create histogram.We can see three categories in below figure .By considering weights , we are able to classify points in three categories.



#### Bivariate Analysis

Sometimes points can be overlapping,You will not be able to classify these categories.So because of these , By using only feature we will not be able to classify three categories. So thats why we go with Bivariate Analysis or Multivariate Analysis.

In Bivariate analysis we take 2 features (Height(y axis) and Weight(X axis)).As we all know that as height increases , weight also increases.But there can be outliers as well , People with less height has high weight.
Based on below figure we are able to categorize the three categories 
If we are able to classify points easily we can use Logistic regression otherwise Non linear models (RF,Decison tree,SVM)



#### Multivariate Analysis

If there are more than 2 features  , we use Multivariate Analysis.By using these you will also get idea of correlation between 2 features.We can easily identify +ve or -ve correlation.If points are evenly distributed between 2 features.Correlation will be 0 for that 2 features.In pearson correlation coefficeint value is betweeen -1 to +1.
If points can be easily classified we can use Linear models otherwise non linear models

## T24 - Univariate , Bivariate , Multivariate Analysis - EDA (Part 2) - Practical

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

df = pd.read_csv('https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv')
df.head()

## We predict the Species category
```

```python
df.shape
```

#### Univariate Analysis

We will take one feature and try to see whether we are able to classify categories of species


```python
df_setosa = df.loc[df['species']=='setosa']
df_virginica = df.loc[df['species']=='virginica']
df_versicolor = df.loc[df['species']=='versicolor']
```

```python
### univaraite analysis on sepal_length (X axis) , Make everythin zero on y axis
plt.plot(df_setosa['sepal_length'],np.zeros_like(df_setosa['sepal_length']),'o')
plt.plot(df_virginica['sepal_length'],np.zeros_like(df_virginica['sepal_length']),'o')
plt.plot(df_versicolor['sepal_length'],np.zeros_like(df_versicolor['sepal_length']),'o')
plt.xlabel('Sepal Length')
plt.show()
```

#### Bivariate Analysis

Analysis will be done on 2 features.

```python
sns.FacetGrid(df,hue='species',height=5).map(plt.scatter,'petal_length','sepal_width').add_legend(); 
plt.show();

## here we can easily classify the categories
```

#### Multivariate Analysis

We will use pairplot

```python
sns.pairplot(df,hue='species',height=3);
```

## T24 - Z-Score Statistics

In Normal Distribution we know it contains some mean and std.
Now if we want to convert Gaussian or Normal Distribution to Standard Normal Distribution (Mean-0,std=1) we apply z score formula.

In Gaussian Dist,We have empirical formula (1st std range - 68%,2nd std range - 95%,3rd std range-99.7%). \
Now i want to find out from 1.5 std away from mean, What will be the distribution.Empirical formula does not help us to find these,So for that we use z score. (Z score table is used)

z_score = (xi - Mean)/std
So now when we apply above formula to {1,2,3,4,5) we get 1 at(-1),2 at(-2),3 at(0),4 at(1),5 at(2).

Z-Score Table Example. (Only Give Left side)

We have data about students in classroom with scores, Mean -75,std-10. \
Now we need to find out Probability where student scores>60

Now when plot these in Gaussian Distribution. Std-10 (Keep that much gap) \



So when we convert Gaussian Distribution to Normal Distribution \
75 will get converted into 0 , 65 - (-1) , 60 - (-1.5).

Now i have to find out what is the probability that student will be scoring > 60 . So we need to find out the filled part in below figure. 
With Empirical formula we will not be able to find out , Thats why we use Z score table



Thar Z score table will tell us what is the distribution below (-1.5) std.Table will contains value about (-1.5).

Now my region will get divided into 3 parts. \
With Z score we will only get 3rd part value - 0.0668. - 6.68%

Now we know that Two parts are symmetric (50-50%) from mean.

Let 1st part pe x ,2nd part  is 50% 

x + 50% + 6.68% = 100 \
